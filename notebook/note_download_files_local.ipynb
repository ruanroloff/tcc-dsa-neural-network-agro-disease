```python
import os 
import boto3
import time
```

```python
#Download files from bucket bucket with reduced files
bucket_name = 'datalake-nessie-agro'
s3_prefix = 'agro/planet/'
```

```python
# Create an S3 client using boto3 and use the paginator to iterate through objects in the specified S3 bucket
s3_client = boto3.client('s3')
paginator = s3_client.get_paginator('list_objects_v2')
pages = paginator.paginate(Bucket=bucket_name, Prefix=s3_prefix)
```

```python
#Local path
local_dir = '/home/sagemaker-user/local/data/'
folder_file_count = 0
```

```python
# Code generated by Generative AI
# Iterate through each page of S3 objects returned by the paginator and donwload locally 
for page in pages:
    if 'Contents' in page:
        for obj in page['Contents']:
            key = obj['Key']
            if key.endswith('/'):
                continue                
                
            #Get the relative path and local path
            relative_path = key[len(s3_prefix):] if key.startswith(s3_prefix) else key
            local_path = os.path.join(local_dir, relative_path)

            # Create any necessary directories in the local path and download the file from S3 to the local path
            os.makedirs(os.path.dirname(local_path), exist_ok=True)
            s3_client.download_file(bucket_name, key, local_path)
            folder_file_count += 1
            
            if folder_file_count % 50 == 0:
                #print(f"Downloaded {folder_file_count} files from {s3_prefix}...")
                print(f"Completed folder: {s3_prefix} - {folder_file_count} files downloaded")
```
